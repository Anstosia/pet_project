## Вывод
### Общая информация



Данные о продажах представлены за период **с 2019-01-31 по 2019-09-30**


* Предоставлено 39 960 записей о продажах.
* **За это время продано 53 874 автомобилей на общую сумму 1.35 млрд USD**
* Осуществлен 1 возврат.
* Наиболее активным месяцем был август. В целом наблюдается растущая тенденция относительно начала года.

Стоимость автомобилей варьируется **от 7 100.0 до 328 510.97 USD**
* В 90% случаях стоимость автомобиля не превышает 49606.35.
* В 90% покупают 1 автомобиль.
* Самое большое число автомобилей в одной продаже 115.

### Популярные бренды и модели
---


На рынке Казазстана значительно выделяется три бренда: **Lada(21%), Toyota(20%), Hyundai(19%).** В натуральном выражени доля рынка данных брендов равна 59%.
Ниже представлены бренды, чья доля рынка в стоимостном выражении  превышает 5% : Toyota(30%), Hyundai(17%), Lada(8%), **Lexus(6%)**.

**Toyota Camry** является самой востребованной машиной, она занимает 9% доли рынка в номинальном выражении и 12% в стоимостном. Для данной модели наиболее популярными являются следующие характеристики: автоматическая коробка передач, **бензиновый** двигатель объемом 2.5, передний привод. Машины 2018 или 2019 года.

В номинальном выражении также популярными являются : Lada Granta(7%) и Hyundai	Tucson(6%).

В стоимостном выражении наибольшую долю рынка, кроме Toyota Camry,занимают: Hyundai	Tucson(6%) и Toyota	LC Prado.

### Характеристики автомобилей
---



* Тип двигателя. Машины на бензиновом двигателе занимают более 90% как в номинальном, так и в стоимостном выражении.
Гибридные и электрокары почти не пользуются спросом. За рассматриваемый период было продано 10 и 13 машин соответственно.

* Страна-производитель. На рынке присутствуют 3 основных страны-производителя: Казахстан, Росиия и Япония. Они занимают 63% рынка в номинальном выражении и 88% в стоимостном.

* Коробка передач. Машины с автоматической коробкой передач занимают 68% рынка. Они приносят 79% прибыли.

* Тип привода. Полный привод занимает большую долю рынка. 39% автомобилей полноприподные и приносят 53% прибыли. Переднеприводные автомобили вторые по пулярности. Они занимают 37% рынка и приносят 26% прибыли.


### Класс машин и сегментация
---

Наибольшая доля рынка в номинальном выражении у машин класса В(26%), они приносят 12% прибыли.
Наибольшую прибыль приносят:  компактные(15%), среднеразмерные(14%) и полноразмерные SUV(13%), а также В(12%) и D(13%) класс.

За весь рассматриваемый период **внедорожники принесли 49% прибыли**. Легковые и коммерческие автомобили вместе занимают 47% рынка. Пикапы и минивены приносят по 2 % от общей прибыли.




### Регион продаж и компании
---
54% прибыли приходится на два региона продаж: Атматы(32%) и Нур-Сутлан(22%)

67% прибыли генерируют 3 компании: Toyota Motor Kazakhstan, Astana Motors и БИПЭК АВТО.

### Проверка гипотезы
---
В рамках исследования была проверена гипотеза: *дорогие и дешевые автомобили в среднем приобретают в одинаковом количестве*

По результатам теста нулевая гипотеза отклонена. Признаем статистически значимую разницу между долями.

Количество купленных дешевых автомобилей больше количества дорогих почти в 2 раза.  

### Заключение
---
Таким образом наиболее предпочитаемый автомобиль покупателя является следующий: на бензиновом двигателе, вероятнее, с автоматической коробкой передач полноприводный или переднеприводный.
Стоимость до 50 тысяч USD.
Чаще всего покупатель физическое лицо.

````python
def info_df(df):
    '''
    Функция для просмотра информации о датафрейме

    '''

    # общая информация о датафрейме
    display(Markdown(f"**Общая информация о датафрейме - {namedf}**"))
    df.info()
    display(Markdown(f"**Первые 10 строк {namedf}**"), df.head(10))

    # проверим возможное количество явных дубликатов
    if df.duplicated().sum() > 0:
        display(Markdown(f"**Количество явных дубликатов:** {df.duplicated().sum()}"))
        display(Markdown(f"**% явных дубликатов:**{round(df.duplicated().sum()/len(df)*100,2)} %"))

    else:
        display(Markdown(f"**Явных дубликатов в {namedf} нет**"))
````

````python
def isnall_info(df):
  '''
  Функция позволяет исследовать пропуски в таблице

  '''
  sns.set_style("whitegrid")
  sns.set_palette("pastel")
  # сформируем таблицу с количеством и процентным соотношением пропусков
  null = df.isnull().agg(['sum', 'mean']).T.sort_values(by='sum', ascending=False).reset_index()
  null.columns = ["Столбец", "Пропуски", "%_пропусков"]
  # приведём значения к %
  null['%_пропусков'] = round(null['%_пропусков']*100, 2)
  # уберем строки, в которых пропуски отсутствуют
  null_bar = null.query("Пропуски > 0")

  # при наличии пропусков выведем информацию о них
  if null_bar.empty == False:
    display(Markdown(f"**Пропуски в {namedf} преддставлены в следующих столбцах:**"))
    display(null_bar)
    # строим барплот для большей наглядности
    sns.barplot(x="Пропуски", y="Столбец", data=null_bar)
    for i, v in enumerate(null_bar['Столбец']):
        plt.text(null_bar['Пропуски'][i], i, str(null_bar['%_пропусков'][i])+"%")

    plt.title('Процент пропусков по столбцам')
    plt.xlabel('Количество пропусков')
    plt.ylabel('Столбец')
    plt.show();

  # в случае отсутсвия пропусков выведем информацию об этом
  else: display(Markdown(f"**Пропуски в {namedf} отсутствуют**"))
````

````python
def change_from_dic(dict_name, name_column, df = df):
    '''
    Функция позволяет заменять данные в столбце по значениям из словаря

    '''
    # произведем замену
    df[name_column] = df[name_column].apply(lambda x: dict_name[x] if x in dict_name else x)
````

````python
def drive_type(cell):
    '''
    Функция сортирует типы приводов
    '''
    if cell in ['4WD', 'quattro', 'Полный', 'полный', 'AWD', '4 WD', '4Motion']:
        return '4WD' #- полный привод
    if cell in ['передний', 'FWD', 'Передний', 'Передний (FF)', 'FF']:
        return 'FWD' #- передний привод
    if cell in ['Задний', 'RWD']:
        return 'RWD' #- задний привод
    if cell in ['4х2.2', '2 WD', '2WD', '4X2', None]:
        return '2WD'
````

````python
def get_client_type(client):
    '''
    функция, которая будет определять тип клиента на основе значения в столбце 'client_type'
    '''

    if isinstance(client, str):
        client = client.lower()
        if 'физ' in client:
            return 'физ.лицо'
        elif 'юр' in client or 'корп' in client:
            return 'юр.лицо'
    return np.nan
````

````python
def pivot_category(category):
    '''
    Функция возвращает сводную таблицу с группировкой признаку.
    Требуется передавать значения для : index = category
    '''
    pivot_table = (df
               .pivot_table(
                  index = category,
                  values=['quantity', 'sale_USD'],
                  aggfunc= 'sum')
               .reset_index())

    pivot_table['sale_USD_m'] = pivot_table['sale_USD']/1000000
    pivot_table = pivot_table.drop(columns = ['sale_USD'],axis = 1)

    # посчитаем процентный вклад каждого значения признака к общему количеству продаж
    pivot_table['persent_of_total_quantity'] = \
    pivot_table['quantity'].apply(lambda x: x/pivot_table['quantity'].sum()*100)

    # посчитаем процентный вклад каждого значения признака к общей сумме продаж
    pivot_table['persent_of_total_sales'] = \
    pivot_table['sale_USD_m'].apply(lambda x: x/pivot_table['sale_USD_m'].sum()*100)
    return pivot_table
````

````python
def info_col_pie (parameter, percent, piv = False, return_df= False):
    '''
    Функция позволяет получить информацию о данных в виде круговой диаграммы
    и сводной таблицы, при необходимости
    '''
    colors = sns.color_palette('pastel')
    # сгруппируем датасет по параметру
    df = pivot_category([parameter]).sort_values(by='sale_USD_m', ascending = False)
    # если требуется вывести сводную таблицу, то задаем аргументу piv значение True
    if piv == True:
        display(df.head(10))
    else: pass

    # Если в таблице не более 5 признаков, то выводим все
    if df.shape[0] <=5:
        parameter_1 = parameter
        parameter_2 = parameter
    # Если в таблице более 5 признаков, то выводим наиболее значимые
    else:
        df['par_quantity'] = (df
                              .apply(lambda x: x[parameter]
                                     if x['persent_of_total_quantity'] >= percent
                                     else 'other', axis=1))
        df['par_revenue'] = (df
                             .apply(lambda x: x[parameter]
                                    if x['persent_of_total_sales'] >= percent
                                    else 'other', axis=1))
        parameter_1 = 'par_quantity'
        parameter_2 = 'par_revenue'


    # формируем две круговые диаграммы
    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.pie(df.groupby(parameter_1)['persent_of_total_quantity'].sum(),
        labels = df.groupby(parameter_1)['persent_of_total_quantity'].sum().index,
        colors = colors,
        autopct='%.0f%%')
    ax1.set_title(f'Доля рынка в натуральном выражении по  {parameter}')


    ax2.pie(df.groupby(parameter_2)['persent_of_total_sales'].sum(),
        labels = df.groupby(parameter_2)['persent_of_total_sales'].sum().index,
        colors = colors,
        autopct='%.0f%%')
    ax2.set_title(f'Доля рынка в стоимостном выражении {parameter}')
    plt.show()

    # при необходимости возвращаем сводную таблицу
    if return_df == True:
        return df
    else: pass
````

#### Изучение пропусков.

Требуется изучить природу пропусков.
* MCAR (Missing Completely At Random / Отсутствует совершенно случайно).В этой ситуации пропуски данных не зависят от наблюдаемых и ненаблюдаемых переменных. Пробелы в данных не ведут к систематической ошибке, но увеличивают случайную, потому что информации становится меньше.

* MAR (Missing At Random) — на практике данные обычно пропущены не случайно, а ввиду некоторых закономерностей. Пропуски относят к MAR, если вероятность пропуска может быть определена на основе другой имеющейся в наборе данных информации (пол, возраст, занимаемая должность, образование…), не содержащей пропуски. В таком случае удаление или замена пропусков на значение «Пропуск», как и в случае MCAR, не приведет к существенному искажению результатов.

* MNAR (Missing Not At Random) — механизм формирования пропусков, при котором данные отсутствуют в зависимости от неизвестных факторов. MNAR предполагает, что вероятность пропуска могла бы быть описана на основе других атрибутов, но информация по этим атрибутам в наборе данных отсутствует. Как следствие, вероятность пропуска невозможно выразить на основе информации, содержащейся в наборе данных.

### Проверка гипотезы

**Гипотеза:**

> Н0 "Дорогие и дешевые автомобили в среднем приобретают в одинаковом количестве"

> Н1 "Количество купленных дешевых и дорогих автомобилей разнится"

Для начала требуется определиться со значением дорогого и дешевого автомобиля.

**Дорогим автомобилем** будем считать тот, чья стоимость равна или превышает 75 процентиль стоимости проданных автомобилей.

**Дешевым** - меньше или равна 25 процентилю.

````python
# определяем стоимость дешевого и дорогого
sale_25 = df['price_usd'].quantile(.25)
sale_75 = df['price_usd'].quantile(.75)
display(Markdown(f'**Стоимость дешевого автомобиля меньше или равна {round(sale_25, 2)}**'))
display(Markdown(f'**Стоимость дорогого автомобиля больше или равна {round(sale_75, 2)}**'))
````

````python
#  сформируем данные о количестве дешевых и дорогих автомобилей.
total = df['quantity'].sum()
ch = df[df['price_usd'] <= sale_25]['quantity']
exp = df[df['price_usd'] >= sale_75]['quantity']
````

````python
# Проведём тест

alpha = .05 # критический уровень статистической значимости

successes = np.array([exp.sum(), ch.sum()])
trials = np.array([total, total])

# пропорция успехов в первой группе:
p1 = successes[0]/trials[0]

# пропорция успехов во второй группе:
p2 = successes[1]/trials[1]

# пропорция успехов в комбинированном датасете:
p_combined = (successes[0] + successes[1]) / (trials[0] + trials[1])

# разница пропорций в датасетах
difference = p1 - p2

# считаем статистику в ст.отклонениях стандартного нормального распределения
z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials[0] + 1/trials[1]))

# задаем стандартное нормальное распределение (среднее 0, ст.отклонение 1)
distr = st.norm(0, 1)

p_value = (1 - distr.cdf(abs(z_value))) * 2

print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
    )
````

````python
print(f'Количество дешевых автомобилей больше количества дорогих в {round(ch.sum()/exp.sum(),2)} раз')
````

По результатам проверки отвергаем нулевую гипотезу. Дешевые автомобили покупают чаще, чем дорогие почти в 2 раза.
